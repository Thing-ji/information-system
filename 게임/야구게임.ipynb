```python
import random

# 컴퓨터 랜덤값 생성, 자리 및 중복 숫자 고려해야함
computer = ['0', '0', '0']
computer[0] = str(random.randrange(1, 9, 1))
computer[1] = str(random.randrange(1, 9, 1))
computer[2] = str(random.randrange(1, 9, 1))
while computer[0] == computer[1]:
    computer[1] = str(random.randrange(1, 9, 1))
while computer[0] == computer[1] or computer[0] == computer[2]:
    computer[2] = str(random.randrange(1, 9, 1))
    
count = 0
strike = 0
ball = 0

while strike < 3:
    user = str(input('세 개의 숫자를 입력하시오. ex)123 : '))
    if user == '':
        print('숫자를 입력하세요.')
        continue
    if len(user) != 3:
        print('세 개의 숫자를 입력하세요.')
        continue
    if user.isalpha():
        print('숫자를 입력해주세요.')
        continue
        
    strike = 0
    ball = 0
    
    for i in range(0, 3):
        for j in range(0, 3):
            if (user[i] == str(computer[j]) and i == j):
                strike += 1
            elif (user[i] == str(computer[j]) and i != j):
                ball += 1
    print('\n[{0}] 스트라이크!, [{1}] 볼! \n'.format(strike, ball))
    count += 1 # 스트라이 카운트
    
print('[{}]번 만에 성공!'.format(count))
print('게임을 종료합니다.')
```
