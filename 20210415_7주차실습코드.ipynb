#### 프레임과 For문을 이용하여 버튼 생성하기

![widget](fig3.png)

```python
from tkinter import *
from decimal import *

# 키 입력 함수:
def click(key):
    entry.insert(END, key)

##### 메인:
window = Tk()
window.title("나의 계산기")

# top_row 프레임 생성
top_row = Frame(window)
top_row.grid(row=0, column=0, columnspan=2, sticky=N)

# 수정 가능한 엔트리 위젯
entry = Entry(top_row, width=45, bg="light green")
entry.grid()

# 숫자 버튼 프레임 생성
num_pad = Frame(window)
num_pad.grid(row=1, column=0, sticky=W)
#num_pad_list = [ '7',  '8',  '9', '4',  '5',  '6', '1',  '2',  '3', '0',  '.',  '=' ]

num_pad_list = [
'7',  '8',  '9',
'4',  '5',  '6',
'1',  '2',  '3',
'0',  '.',  '=' ]

# 반복문으로 숫자 버튼 생성
r = 0 # 행 카운터
c = 0 # 열 카운터
for btn_text in num_pad_list:
    Button(num_pad, text=btn_text, width=5, command=click(btn_text)).grid(row=r,column=c)
    c = c+1
    if c > 2:
        c = 0
        r = r+1

##### 메인 반복문 실행
window.mainloop()
```

#### [실습] 다음의 코드를 생성하여 계산기의 오른쪽에 연산자(operator) 프레임을 생성하고, *, /, +, -, (, ) ,c 버튼을 생성하시오

```python
from tkinter import *
from decimal import *

# 키 입력 함수:
def click(key):
    entry.insert(END, key)

##### 메인:
window = Tk()
window.title("나의 계산기")

# top_row 프레임 생성
top_row = Frame(window)
top_row.grid(row=0, column=0, columnspan=2, sticky=N)

# !!수정!! 가능한 엔트리 위젯
entry = Entry(top_row, width=45, bg="light blue")
entry.grid()

# 숫자 버튼 프레임 생성
num_pad = Frame(window)
num_pad.grid(row=1, column=0, sticky=W)

num_pad_list = [
'7',  '8',  '9',
'4',  '5',  '6',
'1',  '2',  '3',
'0',  '.',  '=' ]

# 반복문으로 숫자 버튼 생성
r = 0 # 행 카운터
c = 0 # 열 카운터
for btn_text in num_pad_list:
    Button(num_pad, text=btn_text, width=5, command=click(btn_text)).grid(row=r,column=c)
    c = c+1
    if c > 2:
        c = 0
        r = r+1

# 연산자 프레임 생성
calcul_pad = Frame(window)
calcul_pad.grid(row = 1, column = 1, sticky = E)

calcul_list = [
'*', '/',
'+', '-',
'(', ')',
'C'
]

# 반복문으로 연산자 버튼 생성
for cal_text in calcul_list:
    Button(calcul_pad, text = cal_text, width = 5, command = click(cal_text)).grid(row=r, column = c)
    c += 1
    if c > 1:
        c = 0
        r += 1

##### 메인 반복문 실행
window.mainloop()
```
